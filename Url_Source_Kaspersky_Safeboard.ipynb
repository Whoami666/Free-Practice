```python
from bs4 import BeautifulSoup
import requests
import re

with open("/tlds-from-a-z.html") as fp:
    soup = BeautifulSoup(fp, "html.parser")

name = "feature-list"
elems = soup.find_all(name)
souper = str(soup.find_all("a", class_ ="smaller80"))
print(souper)

all_small_tlds = re.findall(r">\.[\w|\.]*", souper)
all_small_tlds = [x.strip(">") for x in all_small_tlds]
all_small_tlds[:10]
```

```python
final_tlds_list = all_small_tlds + all_tlds
```

```python
from time import sleep
import sys

def get_the_domen_both(some_address):
  sys.stdout.write(some_address)
  sys.stdout.write('\n')
  sys.stdout.flush()
  sleep(0.25)   

  for i in range(len(final_tlds_list)):
    domen = final_tlds_list[i]
    true_regex = "\\" + str(domen) + '$'
    try:
      if len(re.findall(true_regex, some_address)) > 0:
        only_second = some_address.rstrip(str(domen)).split('.')[-1]
        both = ''.join(only_second) + str(domen)
        break
    except:
      print(f'{domen}, {some_address}') 

  return both
    

def get_the_domen_second(some_address):
  sys.stdout.write(some_address)
  sys.stdout.write('\n')
  sys.stdout.flush()
  sleep(0.25)   

  for domen in final_tlds_list:
    true_regex = "\\" + str(domen) + '$'
    try:
      if len(re.findall(true_regex, some_address)) > 0:
        only_second = some_address.rstrip(str(domen)).split('.')[-1]
        both = ''.join(only_second) + str(domen)
        break
    except:
      print(f'{domen}, {some_address}')
  
  return only_second
```

```python
df_good["second_domen"] = df_good["domen"].apply(get_the_domen_second)
```

```python
df_good["second_plus_tld"] = df_good["domen"].apply(get_the_domen_both)
```
