```python
def two_sum(arr: list, k: int):
    hash_dict = {}
    for i in range(len(arr)):
      if k - arr[i] in hash_dict:
        a = hash_dict[k-arr[i]]
        if i < a:
            return i, a
        else:
            return a, i
      else:
        hash_dict[arr[i]] = i

    return 0, 1


def solution():
    arr = '1 3 4 10'# '5 5 1 3' 
    arr = list(map(int, arr.split()))
    k = 7 #int(input())
    a, b = two_sum(arr, k)
    print(a, b)

solution()
```

```python
def zero_sum_subarrays(arr: list) -> int:
  hash_dict = {0:0}
  counter = 0
  nakop = 0
  for i in arr:
    nakop += i
    if nakop in hash_dict:
      hash_dict[nakop] += 1
      counter += hash_dict[nakop]
    else:
      hash_dict[nakop] = 1
  return counter


def solution():
    inp = '0 -1 1 -1 -1 1 0 -1 -2 2'
    arr = list(map(int, inp.split()))
    count = zero_sum_subarrays(arr)
    print(count)

solution()
```
