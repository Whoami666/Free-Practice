```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_excel('/content/2022 visual (2).xlsx', index_col=0) 
df.head()
```

```python
df['month'] = df['Date'].dt.month
df['month']
```

```python
df = df.set_index('Date')
df = df.drop(index='NaN')
df.head()
```

```python
column_names = [str(i) for i in range(48)] + ['month']
print(column_names)
```

```python
df = df.set_axis(column_names, axis=1)
df.head()
```

```python
group_dict = {}
for name in column_names:
  group_dict[name] = 'value_counts'
df = df.groupby('month').agg(group_dict)
df.head()
```

```python
df_sum = df.sum(axis=1)
df2 = pd.DataFrame(data=df_sum, columns=['activity'])
df2 = df2.apply(lambda x: x/2)
df2.head()
```

```python
df2 = df2.reset_index(level=[0,1]).rename(columns={"level_1": "activity", "activity": "hours"})
df2 = df2[df2['activity'].str.isnumeric() == False]
df2.tail()
```

```python
fig = px.line(df2, x="month", y="hours", color='activity')
fig.show()
```

```python
useless_activity = ["COMMUTING", "FEELING_BAD", "SHOWER", "ORGANIZING", "ORGANIZING OR READING", "ONLINE_SHOPPING", "EATING", "HOUSEWORK", "TRAVEL", "GAMING", "CLASSES", "STUDY_LANG", "SLEEP", "STUDY_UNI", "WTF"]
df_chosen = df2[df2['activity'].isin(useless_activity) == False]
```

```python
fig_chosen = px.line(df_chosen, x="month", y="hours", 
                     color='activity', color_discrete_sequence=px.colors.qualitative.Light24)
fig['data'][3]['line']['color']='black'
fig_chosen.show()
```
